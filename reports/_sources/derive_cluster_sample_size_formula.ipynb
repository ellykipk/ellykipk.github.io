## Sample size calculation

$\textbf{1. For individually randomized clinical trial (rct)}$



First, lets start by  deriving the standard sample size formula for two independent samples z-test

Suppose we want to test 

$$H_0 : \mu_1-\mu_2 =0 \text{ vs } H_1 : \mu_1-\mu_2 >0 $$

Unbiased estimators of $\mu_1 \text{ and } \mu_2$ are $\bar{Y_1}$ and $\bar{Y_2}$ respectively by the weak law of large numbers. 


By the central limit, 

$$\bar{Y_1}-\bar{Y_2} \sim N(\mu_1-\mu_2 , \frac{2\sigma^2}{n} )$$ 

assuming $\bar{Y_1} \text{ and } \bar{Y_2}$ are independent with equal variances. 

Then, we can see that 


$$ Z=\frac{ (\bar{Y_1}-\bar{Y_2}) - (\mu_1-\mu_2)}{\sqrt{\frac{2\sigma^2}{n}} }  \sim N(0 , 1 )$$ 


Under the null(i.e., $H_0$), $\mu_1-\mu_2=0$

$\Rightarrow$

$$\frac{ (\bar{Y_1}-\bar{Y_2})}{\sqrt{\frac{2\sigma^2}{n}} }  \sim N(0 , 1 )$$ 

We reject the null if 

$$\frac{ (\bar{Y_1}-\bar{Y_2})}{\sqrt{\frac{2\sigma^2}{n}} }>C$$ 

where C (some constant) can be choosen such that 

$$ P \left(\frac{ (\bar{Y_1}-\bar{Y_2})}{\sqrt{\frac{2\sigma^2}{n}} }>C \right)=\alpha$$

$\Longleftrightarrow$

$$ P \left(Z>C \right)=\alpha$$

$$ 1-P \left(Z \leq C \right)=\alpha$$

$\Rightarrow$

$$ \Phi^{-1}(1-\alpha)=C $$

or using the more familiar terms

$$ \Phi^{-1}(1-\alpha)=z_{1-\alpha} $$




$\alpha$ being a number between 0 and 1 or the probability of type I error, $Z\sim N(0,1), \Phi(.)$ is the CDF of a normal distribution. 

If one doesn't subscribe to using statistical power for sample size calculation, one can use the fact that we reject the null when 

$$\frac{ (\bar{Y_1}-\bar{Y_2})}{\sqrt{\frac{2\sigma^2}{n}} }>z_{1-\alpha}$$ 

to solve for n in the inequality. 


However, a formal sample size calculation that relies on statistical power goes a step further. 

Notice that we can re-write the rejection region under the  null as 

$$(\bar{Y_1}-\bar{Y_2})>z_{1-\alpha} \sqrt{\frac{2\sigma^2}{n}} $$ 

Under the alternative (i.e., $H_1$), we reject the null if 

$$(\bar{Y_1}-\bar{Y_2})>z_{1-\alpha} \sqrt{\frac{2\sigma^2}{n}} $$ 

But 

$$(\bar{Y_1}-\bar{Y_2}) \sim N(\mu_1-\mu_2, \frac{2\sigma^2}{n} ) $$ 

when the alternative is true. 


Suppose we want to attain a statistical power of $1-\beta$, then we want to calculate the 

$$ P\left( (\bar{Y_1}-\bar{Y_2})>z_{1-\alpha} \sqrt{\frac{2\sigma^2}{n}} \right) $$

And after standardization, we get

$$ P\left( \frac{ (\bar{Y_1}-\bar{Y_2}) - (\mu_1-\mu_2)}{ \sqrt{\frac{2\sigma^2}{n}} } > \frac{z_{1-\alpha} \sqrt{\frac{2\sigma^2}{n}} -  (\mu_1-\mu_2)} {\sqrt{\frac{2\sigma^2}{n}} } \right) $$

Equate it to $1-\beta$, the desired power

$$ P\left( \frac{ (\bar{Y_1}-\bar{Y_2}) - (\mu_1-\mu_2)}{ \sqrt{\frac{2\sigma^2}{n}} } > \frac{z_{1-\alpha} \sqrt{\frac{2\sigma^2}{n}} -  (\mu_1-\mu_2)} {\sqrt{\frac{2\sigma^2}{n}} } \right) =1-\beta $$

Notice that $Z=\frac{ (\bar{Y_1}-\bar{Y_2}) - (\mu_1-\mu_2)}{ \sqrt{\frac{2\sigma^2}{n}} } \sim N(0,1)$ by CLT. Therefore, 

$$ P\left( Z > \frac{z_{1-\alpha} \sqrt{\frac{2\sigma^2}{n}} -  (\mu_1-\mu_2)} {\sqrt{\frac{2\sigma^2}{n}} } \right) =1-\beta $$

$$ 1-P\left( Z \leq \frac{z_{1-\alpha} \sqrt{\frac{2\sigma^2}{n}} -  (\mu_1-\mu_2)} {\sqrt{\frac{2\sigma^2}{n}} } \right) =1-\beta $$


$$P\left( Z \leq \frac{z_{1-\alpha} \sqrt{\frac{2\sigma^2}{n}} -  (\mu_1-\mu_2)} {\sqrt{\frac{2\sigma^2}{n}} } \right)=\beta$$

$\Rightarrow$

$$\frac{z_{1-\alpha} \sqrt{\frac{2\sigma^2}{n}} -  (\mu_1-\mu_2)} {\sqrt{\frac{2\sigma^2}{n}} } = \Phi^{-1}(\beta)$$

Let $\Phi^{-1}(\beta)=z_{\beta}$

Then we get 

$$\frac{z_{1-\alpha} \sqrt{\frac{2\sigma^2}{n}} -  (\mu_1-\mu_2)} {\sqrt{\frac{2\sigma^2}{n}} } = z_{\beta}$$

And after some algebra, and solving for n, we get 

$$n=\frac{2\sigma^2 \left(z_{1-\alpha} - z_\beta \right)^2}{(\mu_1-\mu_2)^2 }$$

Noticing that $-z_\beta=z_{1-\beta}$, we can re-write the equation to 

$$n=\frac{2\sigma^2\left(z_{1-\alpha} + z_{1-\beta} \right)^2}{(\mu_1-\mu_2)^2 }$$

Letting $(\mu_1-\mu_2) =\delta$, we get the more familiar formula 

$$n=\frac{2\sigma^2\left(z_{1-\alpha} + z_{1-\beta} \right)^2}{\delta^2 }$$


$\textbf{2. For cluster randomization}$

Suppose that we have *k* clusters, *m* subjects per cluster for a total of n=mk observations in one treatment group. 

Let 

$$Y_{ijk}=\beta_i+\tau_j+\epsilon_{ijk}$$

where $\beta_i$ is treatment i effect(i=0,1), $\tau_j$ is the random effect of cluster j (j=1,...,b), $\epsilon_{ijk}$ is the random error for subject k (k=1,...,K) in treatment i and cluster j.  $\beta_j \sim N(0, \sigma_{\beta}^2)$, $\epsilon_{ijk} \sim N(0, \sigma_{\epsilon}^2)$,  $\epsilon_{ijk} \text{ and } \beta_j$ are independent.

Let $Y_{ij}$ be cluster-level observation/means. 

We can see that the variance of individual-level response is

$$Var(Y_{ijk})=\sigma^2_{\tau}+\sigma_{\epsilon}^2$$

and the variance of cluster-level response is 

$$Var(Y_{ij})=\sigma^2_{\tau}+ \frac{\sigma_{\epsilon}^2}{m}$$


Suppose we want to test 

$$H_0 : \beta_1-\beta_2 =0 \text{ vs } H_1 : \beta_1-\beta_2 >0 $$

Then $\bar{Y}_1$ and $\bar{Y}_2$ (the mean response by the treatment groups) are the unbiased estimators of $\beta_1$ and $\beta_2 $ respectively. 

What is the variance of $\bar{Y}_1$?

\begin{equation*} 
\begin{split}
Var(\bar{Y}_1)&=Var ( \frac{1}{k}\sum_{i=1}^k Y_{i1}) \\
 &=\frac{1}{k}\times \left( \sigma^2_{\tau}+ \frac{\sigma_{\epsilon}^2}{m}\right) \text{since cluster level responses are independent}
\end{split}
\end{equation*}

It turns out that $\left( \sigma^2_{\tau}+ \frac{\sigma_{\epsilon}^2}{m}\right)$ can be re-written as a function of *k*, *m* and $\rho=\frac{\sigma^2_{\tau}} {\sigma^2_{\tau}+ \sigma_{\epsilon}^2}$. That is 

\begin{equation*} 
\begin{split}
\sigma^2_{\tau}+ \frac{\sigma_{\epsilon}^2}{m}&=\frac{\sigma^2_{\tau}+\sigma_{\epsilon}^2}{m}\left[1+(m-1)\rho\right]\\
&=\frac{\sigma^2_{\tau}+\sigma_{\epsilon}^2}{m} + \frac{(m-1)\sigma^2_{\tau}} {\sigma^2_{\tau}+ \sigma_{\epsilon}^2} \times \frac{\sigma^2_{\tau}+\sigma_{\epsilon}^2}{m}\\
&=\frac{\sigma^2_{\tau}+\sigma_{\epsilon}^2}{m} + \frac{ (m-1)\sigma^2_{\tau}}{m}\\
&= \frac{\sigma^2_{\tau}+\sigma_{\epsilon}^2 + m\sigma^2_{\tau}-\sigma^2_{\tau} } {m} \\
&= \frac{\sigma_{\epsilon}^2 + m\sigma^2_{\tau}}{m}\\
&= \sigma^2_{\tau} + \frac{\sigma_{\epsilon}^2 }{m}
\end{split}
\end{equation*}

Therefore 

\begin{equation*} 
\begin{split}
Var(\bar{Y}_1)&=\frac{1}{k}\times \left( \sigma^2_{\tau}+ \frac{\sigma_{\epsilon}^2}{m}\right)\\
&=\frac{\sigma^2_{\tau}+\sigma_{\epsilon}^2}{mk}\left[1+(m-1)\rho\right]
\end{split}
\end{equation*}

Here, mk=n, the number of subjects in treatment group 1. 

For sample size calculation, the derivation is the same save for the variance component. Here, we replace $\sigma^2$ from individual randomization with $(\sigma^2_{\tau}+\sigma_{\epsilon}^2)\left[1+(m-1)\rho\right]$

In general, we lump $\sigma^2_{\tau}+\sigma_{\epsilon}^2$ together as the population variance and call it $\sigma^2$. Then for sample size calculation, we introduce $\rho$ based on previous data, expert opinion or educated guess. 

Therefore

$$n=\frac{2 (\sigma^2_{\tau}+\sigma_{\epsilon}^2)\left(z_{1-\alpha} + z_{1-\beta} \right)^2}{\delta^2 }\left[1+(m-1)\rho\right]$$



$\large{\textbf{Why way? Individual or cluster rct? }}$

We examine **critical rate of contamination** before the sample size required for individual rct with contamination becomes larger than that of cluster rct. For this, I implement the functions in
*'Hemming, Karla, et al. "Contamination: How much can an individually randomized trial tolerate?." Statistics in Medicine 40.14 (2021): 3329-3351.'*

Cluster randomization is said to be more efficient compared to individual randomization in terms of required sample when 

$$ 2\sigma^2 \left[\frac{(z_{\alpha/2}+z_{\beta}  )^2}{ ((1-\omega)\delta)^2 } \right] >2\sigma^2 \left[\frac{(z_{\alpha/2}+z_{\beta}  )^2}{ \delta^2 } \right] [1+(m-1) \rho] $$

where m is the cluster size and $\rho$ is the intra-cluster correlation (icc). 

For consistency with a real-life clinical setting where the number of doctors are usually fixed, we can convert the formula above to be a function of $k$, the number of clusters as opposed to the number of subjects per cluster (i.e., cluster size). 

Let 

$$n_i=2\sigma^2 \left[\frac{(z_{\alpha/2}+z_{\beta}  )^2}{ \delta^2 } \right]$$ 

Then 

$$\frac{n_i}{(1-\omega)^2} > n_i[1+(m-1) \rho]  $$

$\Longleftrightarrow$

$$ \frac{n_i}{(1-\omega)^2} > n_i \frac{k(1-\rho)}{[k-n_i\rho]}$$

$\Rightarrow$

$$\omega > 1- \sqrt{ \frac{k-n_i\rho}{ k(1-\rho)} } $$

Sample size from cluster randomization is therefore less than the sample size from individual randomization when $\omega > 1- \sqrt{ \frac{k-n_i\rho}{ k(1-\rho)} }$
